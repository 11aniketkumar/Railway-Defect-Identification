```python
from ultralytics import YOLO
```

```python
model = YOLO("best.pt")
```

```python
# predict on image folder ---> caution: OutOfMemory error
model.predict('data//complete_train', save=True, conf=0.8)
```

```python
# cropped prediction on image folder
model.predict('data//spring_assembly', save_crop=True, conf=0.8, project="PROJECT", name="spring_assembly")
```

```python
# model prediction on video
model.predict('data//video//1.mp4', save=True, conf=0.8) #stream=True for OutOfMemory Error
```

```python
import os

images_dir = os.path.join('data','complete_train')

# run model in loop for each image --> prevents OutOfMemory Error
for image in os.listdir(images_dir):
    model.predict(os.path.join(images_dir,image), save=True, conf=0.8)
```

```python
# track the object
model.track(source='test_data',conf=0.8, iou=0.5, save=True, tracker='tracker.yaml')
```
